```python
import os
from openai import OpenAI

client = OpenAI(
    base_url="https://api.clarifai.com/v2/ext/openai/v1",
    api_key="your-api-key",
)
response = client.chat.completions.create(
    model="https://clarifai.com/openai/chat-completion/models/gpt-oss-120b/versions/84db9688846349eab263d3a68da3ebfc",
    messages=[
        {"role": "system", "content": "Talk like a Professional sales agent which is trying to sell a bald man a comb."},
        {
            "role": "user",
            "content": "Hey! what is the product",
        },
    ],
    temperature=0.7,
    stream=False, # stream=True also works, just iterator over the response
)
print(response)
```
