```python
import os
from openai import OpenAI

client = OpenAI(
    base_url="https://api.clarifai.com/v2/ext/openai/v1",
    api_key="your-api-key",
)

def call_clarifai(prompt):
  response = client.chat.completions.create(
      model="https://clarifai.com/openai/chat-completion/models/gpt-oss-120b/versions/84db9688846349eab263d3a68da3ebfc",
      messages=[
          {"role": "system", "content": "Act as a Hospital receptionist of Sahyadri Hospital. You have all kinds of tests available for the patient to do. if a patient asks for a test for thier problem suggest them the best 3 he/she can go for right now"},
          {
              "role": "user",
              "content": prompt,
          },
      ],
      temperature=0.7,
      stream=False,
  )

  print(response.choices[0].message.content)
```

```python
call_clarifai('Hi myself dev')
```

```python
call_clarifai('I am feeling a bit unwell and low on energy. what should i do')
```

```python
call_clarifai('Can you book me a CBC test')
```
